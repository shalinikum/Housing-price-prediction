```python
test_train = test_train.drop(labels = ['Alley','PoolQC','FireplaceQu','Fence','MiscFeature','LotArea'],axis = 1)
test_train.info()
```

```python
test_train.shape
```

```python
column = test_train.columns.tolist()
column.remove('SalePrice')
column
```

```python
for col in column:
    mod = stats.mode(test_train[col])
    test_train[col].replace(np.nan,mod[0][0],inplace = True)
    
test_train.info()
```

```python
categoricals = test_train.select_dtypes(include = ['object'])
#categoricals.remove('MSZoning')
```

```python
# converting categorical variable in numerical column
for cat in categoricals:
    dummy_variable = pd.get_dummies(test_train[cat])
    c = dummy_variable.columns.values
    column={}
    
    for i in range(0,len(c)):
        key = cat + '-'+c[i]
        value = c[i]
        column.update({key:value})
        
    dummy_variable.rename(columns = column)
    test_train = pd.concat([test_train,dummy_variable],axis = 1)
    test_train.drop(cat,axis = 1, inplace = True)
    
test_train.shape
```

```python
train_m = test_train[:1459]
test_m = test_train[1460:]
```

```python
import sklearn
from sklearn.linear_model import LinearRegression
```

```python
column = train_m.columns.values.tolist()
column.remove('SalePrice')
```

```python
lr = LinearRegression()
lr.fit(train_m[column],train_m['SalePrice'])
lr.score(train_m[column],train_m['SalePrice'])
```

```python
yhat = lr.predict(train_m[column])
```

```python
from sklearn.metrics import r2_score
print("R2-score:%.2f" %r2_score(train_m['SalePrice'],yhat))
```
